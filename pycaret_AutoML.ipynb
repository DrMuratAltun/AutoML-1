```python
import streamlit as st
from pycaret.classification import setup, compare_models, pull, save_model
import pandas as pd
import os

# Dataset kontrolü
if os.path.exists('./dataset.csv'):
    df = pd.read_csv('dataset.csv', index_col=None)  # Dosya adı düzeltilmiş.

with st.sidebar:
    st.image('https://leilaabdel.com/img/deep_learning_course_pic.png')
    st.title('AutoML Classification')
    choice = st.radio('Navigation', ['Upload', 'EDA', 'Modelling', 'Download'])  # 'Navigaiton' yazım hatası düzeltildi.

if choice == 'Upload':
    st.title('Upload your data file')
    file = st.file_uploader('Upload your data')
    if file:
        df = pd.read_csv(file, index_col=None)
        df.to_csv('dataset.csv', index=None)  # 'dataset1.csv' dosya adı düzeltilmiş.
        st.dataframe(df.head())

if choice == 'EDA':
    st.title('Exploratory Data Analysis')
    if 'df' in locals():
        profile_df = df.profile_report()  # 'pandas_profiling' kütüphanesi yerine yüklenmeli.
        from streamlit_pandas_profiling import st_profile_report
        st_profile_report(profile_df)

if choice == 'Modelling':
    if 'df' in locals():
        target_col = st.selectbox('Choose the target column', df.columns)
        if st.button('Train model'):  # 'dt' yerine 'st' kullanılmalı.
            setup(data=df, target=target_col)
            best_model = compare_models()
            compare_df = pull()
            st.dataframe(compare_df)
            save_model(best_model, 'best_model.pkl')  # Fonksiyon ataması hatası düzeltilmiş.

if choice == 'Download':
    try:
        with open('best_model.pkl', 'rb') as f:
            st.download_button('Download the best model', f, 'best_model.pkl')
    except Exception as e:
        st.error(f"Error downloading the model: {str(e)}")
```
